```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
from pandas_datareader import data as pdr
import yfinance as yfin
yfin.pdr_override()
```

```python
end_date = dt.datetime.now()
start_date = end_date - dt.timedelta(days=300)
```

```python
end_date, start_date
```

```python
def get_data(stocks, start_date, end_date):
    stockData = pdr.get_data_yahoo(stocks, start_date, end_date)
    stockData = stockData['Close']
    returns = stockData.pct_change()
    meanReturns = returns.mean()
    covMatrix = returns.cov()
    return meanReturns, covMatrix
    
```

```python
stockList = ['CBA', 'BHP', 'TLS', 'NAB', 'WBC', 'STO']
stocks = [stock + '.AX' for stock in stockList]
mean_returns, cov_matrix = get_data(stocks, start_date, end_date)
```

```python
print(mean_returns)
```

```python
weigths = np.random.random(len(mean_returns)) / len(mean_returns)
```

```python
weigths
```

```python
plt.figure(figsize=(15, 9))
mc_sims = 1000
T = 100
meanM = np.full(shape=(T, len(weigths)), fill_value=mean_returns)
```

```python
meanM
```

```python
meanM.shape
```

```python
mc_sims
```

```python
meanM = meanM.T
portfolio_sims = np.full(shape=(T, mc_sims), fill_value=0.0)
initialPortfolio=1000
for m in range(0, mc_sims):
    Z = np.random.normal(size = (T, len(weigths)))
    L = np.linalg.cholesky(cov_matrix)
    dailyReturns = meanM + np.inner(L, Z)
    portfolio_sims[:, m] = np.cumprod(np.inner(weigths, dailyReturns.T) + 1) * initialPortfolio
plt.plot(portfolio_sims)
plt.ylabel('Portfolio Value($)')
plt.xlabel('Days')
plt.title('MC sims')
plt.show()
```

```python
def mcVaR(returns, alpha=5):
    if isinstance(returns, pd.Series):
        return np.percentile(returns, alpha)
    else:
        raise TypeError("Expected pandas series")
    
```

```python
def mcCVaR(returns, alpha=5):
    if isinstance(returns, pd.Series):
        below_CVaR = returns <= mcVaR(returns, alpha)
        return returns[below_CVaR].mean()
    else:
        raise TypeError("Expected pandas series")

portfolioResults = pd.Series(portfolio_sims[-1, :])
VaR = initialPortfolio - mcVaR(portfolioResults, alpha=5)
CVaR = initialPortfolio - mcCVaR(portfolioResults, alpha=5)

print("VaR = {}".format(round(VaR, 2)))
print("CVaR = {}".format(round(CVaR, 2)))
      
```

```python
import datetime

S = 101.95
K = 98.15
vol = 0.0991
r = 0.01
N = 10
M = 10000

market_value = 3.86
T = ((datetime.date(2023, 3, 17) - datetime.date(2022, 12, 17)).days + 1) / 365
```

```python
T
```

```python
dt = T / N
nu_dt = (r - 0.5 * vol ** 2) * dt
vol_sdt = vol * np.sqrt(dt)
lnS = np.log(S)

sum_CT = 0
sum_CT2 = 0

for i in range(M):
    lnSt = lnS
    for j in range(N):
        lnSt += nu_dt + vol_sdt * np.random.normal()
    ST = np.exp(lnSt)
    CT = max(0, ST - K)
    sum_CT += CT
    sum_CT2 += CT * CT

# Compute expectation and SE
C0 = (np.exp(-r * T) * sum_CT) / M
sigma = np.sqrt((sum_CT2 - sum_CT * sum_CT / M) * np.exp(-2 * r * T / (M - 1)))
SE = sigma / np.sqrt(M)
print("Call value is ${0} with SE +/- {1}".format(np.round(C0, 2), np.round(SE, 2)))
```

```python
# MC Method
Z = np.random.normal(size=(N, M))
Z
```

```python
# ?? delta_lnSt lnSt = ln(S) + list of delta_t following a normal law
delta_lnSt = nu_dt + vol_sdt * Z
```

```python
lnSt = lnS + np.cumsum(delta_lnSt, axis=0)
lnSt = np.concatenate((np.full(shape=(1, M), fill_value=lnS), lnSt))

# compute expectation and SE
ST = np.exp(lnSt)
CT = np.maximum(0, ST - K)
C0 = np.exp(-r*T)*np.sum(CT[-1])/M

sigma = np.sqrt(np.sum((CT[-1] - C0) ** 2)/(M-1))
SE = sigma/np.sqrt(M)
```
